Просчет метода Рунге – Кутта 4 порядка для модели эпидемии

```python
# Данные модели
a = float(input("Введите коэффициент а, скорость убывания здоровых особей: "))
b = float(input("Введите коэффициент b, скорость роста заболевания: "))
h = int(input("Введите шаг: "))
days = int(input("Введите количество дней для расчета: "))

# Начальные условия
h0 = int(input("Введите количество инфицированных особей: "))
g0 = int(input("Введите количество восприимчивых особей: "))
z0 = int(input("Введите количество невосприимчивых особей: "))

res = [(0, h0, g0, z0)]

for d in range(1, days + 1):
    # k1
    kh1 = h * (a * h0 * g0 - b * h0)
    kg1 = h * (-a * h0 * g0)
    kz1 = h * (b * h0)

    # k2
    kh2 = h * (a * (h0 + 0.5 * h * kh1) * (g0 + 0.5 * h * kg1) - b * (h0 + 0.5 * h * kh1))
    kg2 = h * (-a * (h0 + 0.5 * h * kh1) * (g0 + 0.5 * h * kg1))
    kz2 = h * (b * (h0 + 0.5 * h * kh1))

    # k3
    kh3 = h * (a * (h0 + 0.5 * h * kh2) * (g0 + 0.5 * h * kg2) - b * (h0 + 0.5 * h * kh2))
    kg3 = h * (-a * (h0 + 0.5 * h * kh2) * (g0 + 0.5 * h * kg2))
    kz3 = h * (b * (h0 + 0.5 * h * kh2))

    # k4
    kh4 = h * (a * (h0 + h * kh3) * (g0 + h * kg3) - b * (h0 + h * kh3))
    kg4 = h * (-a * (h0 + h * kh3) * (g0 + h * kg3))
    kz4 = h * (b * (h0 + h * kh3))

    # Значения (метод Рунге-Кутты 4-го порядка)
    h0 = h0 + (1/6) * (kh1 + 2 * kh2 + 2 * kh3 + kh4)
    g0 = g0 + (1/6) * (kg1 + 2 * kg2 + 2 * kg3 + kg4)
    z0 = z0 + (1/6) * (kz1 + 2 * kz2 + 2 * kz3 + kz4)

    # Сохраняем результаты
    res.append((d, h0, g0, z0))

# Вывод результатов
print("\nРезультаты моделирования:")
for d, h_val, g_val, z_val in res:
    print(f"День {d}: h = {h_val:.4f}, g = {g_val:.4f}, z = {z_val:.4f}")
```
