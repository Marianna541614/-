```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
# Параметры модели
a = float(input('Введите коэффициент а, скорость убывания здоровых особей'))
b = float(input("Введите численность b, скорость роста заболевших"))
n = int(input("Введите численность популяции"))
# Начальные условия
x_0 = int(input())
y_0 = int(input())
z_0 = int(input())

# Функция для системы уравнений
def model (y, t, a, b):
  x, y = y
  dxdt = a * x * y - b * x
  dydt = -a * x * y
  return [dxdt, dydt]

t = np.linspace(0, 100, 1000) # Время
#Решение системы
intilian_conditions = [x_0, y_0]
solution = odeint(model, intilian_conditions, t , args=(a,b))
x = solution[:, 0]
y = solution[:, 1]
z = n- x - y

#Построение графика
plt. figure( figsize =(10,6))
plt.plot(t, x, label = "Инфицированные (х)")
plt.plot(t, y, label = "Восприимчивые (у)")
plt.plot(t, z, label = "Невосприимчивые (z)")
plt.xlabel('Время')
plt.ylabel('Численность')
plt.legend()
plt.title('Моделирование распространения заболевания')
plt.show()
```
