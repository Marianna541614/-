Просчет метода Рунге – Кутта 4 порядка для модели эпидемии с учетом
рождаемости и смертности

```python
# Данные модели
a = float(input("Введите коэффициент a, скорость убывания здоровых особей: "))
b = float(input("Введите коэффициент b, скорость роста заболевания: "))
c = float(input("Введите коэффициент c, коэффициент смертности: "))
N = int(input("Введите общую численность: "))

# Начальные условия
h0 = int(input("Введите количество инфицированных особей: "))
g0 = int(input("Введите количество восприимчивых особей: "))
z0 = int(input("Введите количество умерших или переболевших особей: "))

# Время
t0 = 0
tf = int(input("Введите количество дней для расчета: "))
h = int(input("Введите шаг: "))

res = [(0, h0, g0, z0)]

for n in range(1, tf + 1):
    # k1
    kh1 = h * (a * h0 * g0 - b * h0 - c * h0)
    kg1 = h * (-a * h0 * g0 + c * N - c * g0)
    kz1 = h * (b * h0 - c * z0)

    # k2
    kh2 = h * (a * (h0 + 0.5 * h * kh1) * (g0 + 0.5 * h * kg1) -
               b * (h0 + 0.5 * h * kh1) - c * (h0 + 0.5 * h * kh1))
    kg2 = h * (-a * (h0 + 0.5 * h * kh1) * (g0 + 0.5 * h * kg1) +
               c * N - c * (g0 + 0.5 * h * kg1))
    kz2 = h * (b * (h0 + 0.5 * h * kh1) - c * (z0 + 0.5 * h * kz1))

    # k3
    kh3 = h * (a * (h0 + 0.5 * h * kh2) * (g0 + 0.5 * h * kg2) -
               b * (h0 + 0.5 * h * kh2) - c * (h0 + 0.5 * h * kh2))
    kg3 = h * (-a * (h0 + 0.5 * h * kh2) * (g0 + 0.5 * h * kg2) +
               c * N - c * (g0 + 0.5 * h * kg2))
    kz3 = h * (b * (h0 + 0.5 * h * kh2) - c * (z0 + 0.5 * h * kz2))

    # k4
    kh4 = h * (a * (h0 + h * kh3) * (g0 + h * kg3) -
               b * (h0 + h * kh3) - c * (h0 + h * kh3))
    kg4 = h * (-a * (h0 + h * kh3) * (g0 + h * kg3) +
               c * N - c * (g0 + h * kg3))
    kz4 = h * (b * (h0 + h * kh3) - c * (z0 + h * kz3))

    # Значения (метод Рунге-Кутты 4-го порядка)
    h0 = h0 + (1/6) * (kh1 + 2 * kh2 + 2 * kh3 + kh4)
    g0 = g0 + (1/6) * (kg1 + 2 * kg2 + 2 * kg3 + kg4)
    z0 = z0 + (1/6) * (kz1 + 2 * kz2 + 2 * kz3 + kz4)

    # Сохраняем результаты
    res.append((n, h0, g0, z0))

# Вывод результатов
print("\nРезультаты моделирования:")
for n, h_val, g_val, z_val in res:
    print(f"День {n}: h = {h_val:.4f}, g = {g_val:.4f}, z = {z_val:.4f}")
```
